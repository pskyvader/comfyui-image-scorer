# 2D Parameter Analysis and Visualization

This notebook demonstrates the `ParameterAnalyzer` class for analyzing image scoring data with scatter plot visualization and correlation analysis.

```python
import sys
from pathlib import Path

# Add project root to path
project_root = Path.cwd()
sys.path.insert(0, str(project_root))

print(f'Project root: {project_root}')
```

```python
from shared.training.parameter_analysis import ParameterAnalyzer
import json

print('✓ ParameterAnalyzer imported successfully')
```

## Load Sample Data

Load vectors and scores data from JSONL files for analysis.

```python
from shared.io import load_single_jsonl
from shared.paths import vectors_file, scores_file

print(f'Vectors file: {vectors_file}')
print(f'Scores file: {scores_file}')

# Load data (limit to first 100 for faster processing)
vectors_data = load_single_jsonl(vectors_file)[:100]
scores_data = load_single_jsonl(scores_file)[:100]

print(f'\n✓ Loaded {len(vectors_data)} vector records')
print(f'✓ Loaded {len(scores_data)} score records')

# Show sample record
if vectors_data:
    print(f'\nSample vector record keys: {list(vectors_data[0].keys())[:5]}')
if scores_data:
    print(f'Sample score record: {scores_data[0]}')
```

## Initialize ParameterAnalyzer

Create an analyzer instance to compute correlations and generate visualizations.

```python
# Initialize analyzer
analyzer = ParameterAnalyzer(vectors_data, scores_data)
print('✓ ParameterAnalyzer initialized')

# Compute correlations
print('\nComputing correlations...')
correlations = analyzer.compute_correlations()
print(f'✓ Computed correlations: {len(correlations)} dimensions analyzed')

# Show top correlations
print('\nTop 5 correlations with score:')
top_corr = sorted(correlations.items(), key=lambda x: abs(x[1]), reverse=True)[:5]
for dim, corr in top_corr:
    print(f'  Dimension {dim}: {corr:.4f}')
```

## Generate Scatter Plots

Create visualization of the top correlated parameters.

```python
# Generate scatter plots for top 4 correlated dimensions
print('Generating scatter plots...')
try:
    plots = analyzer.generate_scatter_plots(output_dir='parameter_analysis', num_plots=4)
    print(f'✓ Generated {len(plots)} scatter plots')
    for plot in plots:
        print(f'  - {plot}')
except Exception as e:
    print(f'Note: Scatter plot generation requires matplotlib: {e}')
```

## Generate Summary Report

Create a JSON report with correlation statistics.

```python
# Generate summary report
print('Generating summary report...')
report = analyzer.generate_summary_report()
print(f'✓ Report generated')

# Display report contents
print('\nReport structure:')
for key in report.keys():
    if isinstance(report[key], dict):
        print(f'  {key}: {list(report[key].keys())[:5]}...')
    elif isinstance(report[key], list):
        print(f'  {key}: list of {len(report[key])} items')
    else:
        print(f'  {key}: {type(report[key]).__name__}')
```

## Save Report

Save the analysis report to a JSON file.

```python
# Save report
from pathlib import Path
import json

output_path = Path('parameter_analysis/analysis_report.json')
output_path.parent.mkdir(parents=True, exist_ok=True)

with open(output_path, 'w') as f:
    json.dump(report, f, indent=2)

print(f'✓ Report saved to {output_path}')
print(f'✓ File size: {output_path.stat().st_size} bytes')
```
